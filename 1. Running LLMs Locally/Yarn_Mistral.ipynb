# **Yarn Mistral**

Thi model has a context window of 128k tokens
This is made from the mistral base model hence it's a model that works by predicting the next word. It is not **fined tuned** because is a modified version of *mistral 7b base* model. This means you can not chart or have conversation whith the Yan Mistral model but rather have text completion

installing the required dependency

````python
#importing AutoModelForCausalLM from ctranformers in order to downlad the model from huggingface
from ctransformers import AutoModelForCausalLM

# Set gpu_layers to the number of layers to offload to GPU. Set to 0 if no GPU acceleration is available on your system.
llm = AutoModelForCausalLM.from_pretrained("TheBloke/Yarn-Mistral-7B-128k-GGUF", model_file="yarn-mistral-7b-128k.Q4_K_M.gguf", model_type="mistral", gpu_layers=50)

#prompting the model for text completion in print(llm("Your text here"))
print(llm("Raila Amolo Odinga"))
````
